```python
import numpy as np
import pandas as pd
```

```python
TR1_list1=pd.read_csv("/content/i1.csv")
TR2_list2=pd.read_csv("/content/i2.csv")
```

```python
Data=TR1_list1.merge(TR2_list2, how='outer')
Data
```

```python
Data.shape
```

```python
Data.info()
```

```python
Data.describe()
```

```python
Data=Data.rename(columns={'Invoice No.':'Invoice','Qty.':'Quantity','Table No.':'Table_No'})
```

```python
Data=Data[Data["Final Total"]>0]
```

```python
Data=Data[Data["Price"]>0]
```

```python
Data=Data[Data["Tax"]>0]
```

```python
Data=Data[Data["Quantity"]>0]
```

```python
Data.isnull().sum()
```

```python
Data.drop(['Variation','HSN'],axis=1,inplace=True)
```

```python
Data.shape
```

```python
Data.info()
```

```python
Data['Date']=pd.to_datetime(Data.Date,format='%Y-%m-%d')
```

```python
Data['Date']
```

```python
xyz=[]
for i in Data.columns:
  if(Data[i].dtypes=="object"):
    xyz.append(i)
print(xyz)

```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Label=LabelEncoder()
for i in xyz:
  Data[i]=Data[[i]].apply(Label.fit_transform)
```
